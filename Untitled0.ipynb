```python
to_path = r"/content/drive/MyDrive/cow_detection_new/images"

def move_file():
    n = 1
    folders = os.listdir(os.getcwd())[1:-1]
    folders.sort()
    for folder in folders:
        #print(folder)
        path = os.path.join(os.getcwd(), folder)
        files = os.listdir(path)
        files.sort()
        #print(files)
        #print()
        #continue
        for file in files:
            with open(os.path.join(path,file), 'rb') as f:
                m = "{0:04d}.JPG".format(n)
                pic = f.read()
                dest_pic = open(to_path+"/DJI_"+m, 'wb')
                dest_pic.write(pic)
                dest_pic.close()
                n += 1
```

```python
move_file()
```
