```python
import torch
#from IPython.display import Image
import os
import random
from tqdm import tqdm
from PIL import Image, ImageDraw
import numpy as np
import matplotlib.pyplot as plt
```

```python
def extract_info_from_txt(txt_file):
    fnum = 1
    n_channels = 3
    width, height = 4000, 3000
    files = []
    
    with open(txt_file, 'r') as f:
        info = f.readlines()
    for line in info:
        info_dict = {}
        info_dict['bboxes'] = []
        info_dict['image_size'] = (width, height, n_channels)
        l = line.split()
        info_dict['filename'] = "DJI_{0:04d}.JPG".format(fnum)
        fnum += 1
        for coord in l[2:]:
            xmin, ymin, xmax, ymax = tuple(map(float, coord.split(',')))
            bbox = {'class':'animal', 'xmin':xmin, 'ymin':ymin, 'xmax':xmax, 'ymax':ymax}
            info_dict['bboxes'].append(bbox)
        files.append(info_dict)
    return files        
```

```python
class_name_to_id = {'animal':0}
```

```python
def convert_to_yolov5(files):
    for info in files:
        print_buffer = []
        for bbox in info['bboxes']:
            try:
                class_id = class_name_to_id[bbox['class']]
            except KeyError:
                print("Invalid class, must be one from ", class_name_to_id.keys())
            # Transform coordinate to format expected by yolov5
            b_center_x = (bbox["xmin"] + bbox["xmax"])/2
            b_center_y = (bbox["ymin"] + bbox["ymax"])/2
            b_width = (bbox["xmax"] - bbox["xmin"])
            b_height = (bbox["ymax"] - bbox["ymin"])
            
            # Normalize the coordinates by the dimension of the image
            image_w, image_h, image_c = info["image_size"]
            b_center_x /= image_w
            b_center_y /= image_h
            b_width /= image_w
            b_height /= image_h
            
            # write bounding box to file
            print_buffer.append("{} {:.3f} {:.3f} {:.3f} {:.3f}".format(class_id, b_center_x, b_center_y, b_width, b_height))
            
        save_fname = os.path.join("annotations", info["filename"].replace(".JPG", ".txt"))
        
        print("\n".join(print_buffer), file=open(save_fname, "w"))
```

```python
random.seed(0)
import cv2
def plot_bounding_box(image, annotation_list):
    fig = plt.figure(figsize=(20,15))
    annotations = np.array(annotation_list, dtype=np.float64)
    w, h = tuple(map(float, image.size))
    
    plotted_image = ImageDraw.Draw(image)
    
    transformed_annotations = np.copy(annotations)
    transformed_annotations[:, [1,3]] = annotations[:, [1,3]] * w
    transformed_annotations[:, [2,4]] = annotations[:, [2,4]] * h
    
    transformed_annotations[:, 1] = transformed_annotations[:, 1] - (transformed_annotations[:, 3]/2)
    transformed_annotations[:, 2] = transformed_annotations[:, 2] - (transformed_annotations[:, 4]/2)
    transformed_annotations[:, 3] = transformed_annotations[:, 1] + transformed_annotations[:, 3]
    transformed_annotations[:, 4] = transformed_annotations[:, 2] + transformed_annotations[:, 4]
    
    for ann in transformed_annotations:
        obj_cls, x0, y0, x1, y1 = ann
        plotted_image.rectangle(((x0,y0),(x1,y1)), outline='red', width=10)
        plotted_image.text((x0, y0-10), 'animal')
        
    plt.imshow(np.array(image))
    plt.savefig("test.jpg")
```
